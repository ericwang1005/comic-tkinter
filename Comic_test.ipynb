### 找出搜尋後資料

```python
soup=BeautifulSoup(chrome.page_source,'lxml')
soup
```

```python
book_list=soup.find('ul',class_="book-list").find_all('li')
book_list
```

```python
book_list[0]
```

```python
datas=[]
for book in book_list:
    link='https://www.manhuaren.com'+book.find('a').get('href')
    title=book.find('a').get('title')
    info=book.find('p',class_="book-list-info-desc").text.strip()
    image_url=book.find('img').get('src')
    
    datas.append([title,info,link,image_url])
    
datas
```

### 輸出圖片等連結

```python
df=pd.DataFrame(datas,columns=['title','info','link','image_url'])
df
```

```python
df['image_url']
```

### 模擬tkinter選單
